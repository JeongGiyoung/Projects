```python
path_detectedFolder = './runs/detect/' 
file_list = os.listdir(path_detectedFolder) 

print (file_list)
```

# # OCR product logo

```python
import easyocr
```

```python
reader = easyocr.Reader(['en'])
```

```python
path_croppedImage = f'{path_detectedObject}/{detectedObject[0]}/'
path_croppedImage
```

```python
list_croppedImage = os.listdir(path_croppedImage) 

print (list_croppedImage)
```

```python
for idx in list_croppedImage :
    tmp = f'{path_croppedImage}{idx}'
    result = reader.readtext(tmp, detail = 0)
    print(result)
```
